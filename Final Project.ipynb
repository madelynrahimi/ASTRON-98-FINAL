```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
raw_frame = fits.open('SOFI.2020-03-06T01_07_42.744.fits')
raw_frame
```

```python
image_data_rf = raw_frame[0].data
print(image_data_rf)
```

```python
plt.figure()
plt.imshow(image_data_rf, origin = "lower", cmap = "inferno", vmin = 400, vmax = 5000)
plt.colorbar()
plt.title("Raw Image (Figure 1)")
plt.show()
```

```python
flat_field_1 = fits.open('SOFI.2020-03-05T13_16_15.529.fits')
flat_field_2 = fits.open('SOFI.2020-03-05T13_17_56.530.fits')
flat_field_3 = fits.open('SOFI.2020-03-05T14_15_37.923.fits')
flat_field_4 = fits.open('SOFI.2020-03-05T13_35_11.211.fits')
```

```python
image_data_ff1 = flat_field_1[0].data
image_data_ff2 = flat_field_2[0].data
image_data_ff3 = flat_field_3[0].data
image_data_ff4 = flat_field_4[0].data
print(image_data_ff1)
print(image_data_ff2)
print(image_data_ff3)
print(image_data_ff4)
```

```python
fig, axs = plt.subplots(2,2)
ff1_plot = axs[0,0].imshow(image_data_ff1, origin = "lower", cmap = "inferno")
axs[0,0].set_xlabel('Flat Frame 1')
fig.colorbar(ff1_plot, ax=axs[0, 0], location='right')
ff2_plot = axs[0,1].imshow(image_data_ff2, origin = "lower", cmap = "inferno")
axs[0,1].set_xlabel('Flat Frame 2')
fig.colorbar(ff2_plot, ax=axs[0, 1], location='right')
ff3_plot = axs[1,0].imshow(image_data_ff3, origin = "lower", cmap = "inferno")
axs[1,0].set_xlabel('Flat Frame 3')
fig.colorbar(ff3_plot, ax=axs[1, 0], location='right')
ff4_plot = axs[1,1].imshow(image_data_ff4, origin = "lower", cmap = "inferno")
axs[1,1].set_xlabel('Flat Frame 4')
fig.colorbar(ff4_plot, ax=axs[1, 1], location='right')
plt.suptitle("Flat Frame Images (Figure 2)")
plt.tight_layout()
plt.show()
```

```python
dark_frame_1 = fits.open('SOFI.2020-03-05T19_25_26.001.fits')
dark_frame_2 = fits.open('SOFI.2020-03-05T19_26_53.806.fits')
dark_frame_3 = fits.open('SOFI.2020-03-05T19_28_21.734.fits')
dark_frame_4 = fits.open('SOFI.2020-03-05T19_59_37.456.fits')
```

```python
image_data_df1 = dark_frame_1[0].data
image_data_df2 = dark_frame_2[0].data
image_data_df3 = dark_frame_3[0].data
image_data_df4 = dark_frame_4[0].data
print(image_data_df1)
print(image_data_df2)
print(image_data_df3)
print(image_data_df4)
```

```python
fig, axs = plt.subplots(2,2)
df1_plot = axs[0,0].imshow(image_data_df1, origin = "lower", cmap = "twilight", vmin = 0, vmax = 1000)
axs[0,0].set_xlabel('Dark Frame 1')
fig.colorbar(df1_plot, ax=axs[0, 0], location='right')
df2_plot = axs[0,1].imshow(image_data_df2, origin = "lower", cmap = "twilight", vmin = 0, vmax = 1000)
axs[0,1].set_xlabel('Dark Frame 2')
fig.colorbar(df2_plot, ax=axs[0, 1], location='right')
df3_plot = axs[1,0].imshow(image_data_df3, origin = "lower", cmap = "twilight", vmin = 0, vmax = 1000)
axs[1,0].set_xlabel('Dark Frame 3')
fig.colorbar(df3_plot, ax=axs[1, 0], location='right')
df4_plot = axs[1,1].imshow(image_data_df4, origin = "lower", cmap = "twilight", vmin = 0, vmax = 1000)
axs[1,1].set_xlabel('Dark Frame 4')
fig.colorbar(df4_plot, ax=axs[1, 1], location='right')
plt.suptitle("Dark Frame Images (Figure 3)")
plt.tight_layout()
plt.show()
```

```python
average_ff_arr = (np.array(image_data_ff1)+ np.array(image_data_ff2)+ np.array(image_data_ff3)+ np.array(image_data_ff4))/4
print(average_ff_arr)
```

```python
plt.figure()
plt.imshow(average_ff_arr, origin = "lower", cmap = "inferno")
plt.colorbar()
plt.title('Averaged Flat Frame Image (Figure 4)')
plt.show()
```

```python
average_df_arr = (np.array(image_data_df1)- np.array(image_data_df2)+ np.array(image_data_df3)+ np.array(image_data_df4))/4
print(average_df_arr)
```

```python
plt.figure()
plt.imshow(average_df_arr, origin = "lower", cmap = "twilight", vmin = 0, vmax = 1000)
plt.colorbar()
plt.title('Averaged Dark Frame Image (Figure 5)')
plt.show()
```

```python
final_frame = (np.array(image_data_rf)- np.array(average_df_arr))/np.array(average_ff_arr)
print(final_frame)
```

```python
plt.figure()
plt.imshow(final_frame, origin = "lower", cmap = "inferno", vmin = 0, vmax = 2)
plt.colorbar()
plt.title("Final Image (Figure 6)")
plt.show()
```
